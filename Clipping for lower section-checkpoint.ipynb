```python
import rasterio
import geopandas as gpd
from rasterio.mask import mask
import os

# Define file paths
geotiff_path = r"E:\InSAR Paper data\Mosaic def\mosaic_def.tif"
shapefile_path = r"E:\InSAR Paper data\Road QTH (lower section)\road_clipped_utm46n.shp"
output_folder = r"E:\InSAR Paper data\Mosaic deformation"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Define output file path
output_path = os.path.join(output_folder, "mosaic_def_clipped.tif")

# Read the shapefile
print("Reading shapefile...")
shapefile = gpd.read_file(shapefile_path)

# Read the GeoTIFF
print("Reading GeoTIFF...")
with rasterio.open(geotiff_path) as src:
    print(f"GeoTIFF CRS: {src.crs}")
    print(f"Shapefile CRS: {shapefile.crs}")
    print(f"GeoTIFF bounds: {src.bounds}")
    
    # Reproject shapefile to match GeoTIFF's CRS
    print(f"Reprojecting shapefile to match GeoTIFF CRS: {src.crs}...")
    shapefile_reprojected = shapefile.to_crs(src.crs)
    
    print(f"Reprojected shapefile bounds: {shapefile_reprojected.total_bounds}")
    
    # Get the geometry of the reprojected shapefile
    geometries = shapefile_reprojected.geometry.values
    
    # Clip the raster using the reprojected shapefile geometry
    print("Clipping GeoTIFF to shapefile extent...")
    try:
        out_image, out_transform = mask(src, geometries, crop=True, nodata=src.nodata)
        
        # Update metadata for the clipped raster
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_image.shape[1],
            "width": out_image.shape[2],
            "transform": out_transform,
            "nodata": src.nodata
        })
        
        # Save the clipped GeoTIFF
        print(f"Saving clipped GeoTIFF to {output_path}")
        with rasterio.open(output_path, "w", **out_meta) as dest:
            dest.write(out_image)
            
        print(f"Clipping complete! Clipped GeoTIFF saved to: {output_path}")
        
        # Print basic information about the clipped raster
        print(f"\nClipped GeoTIFF information:")
        print(f"Dimensions: {out_image.shape[2]} x {out_image.shape[1]} pixels")
        print(f"Bounds: {rasterio.transform.array_bounds(out_image.shape[1], out_image.shape[2], out_transform)}")
        
    except ValueError as e:
        print(f"Error: {e}")
        print("\nTrying alternative approach - using buffer around shapefile...")
        
        # Try buffering the shapefile slightly to ensure intersection
        shapefile_buffered = shapefile_reprojected.copy()
        shapefile_buffered.geometry = shapefile_buffered.geometry.buffer(10)  # 10 unit buffer
        geometries = shapefile_buffered.geometry.values
        
        out_image, out_transform = mask(src, geometries, crop=True, nodata=src.nodata)
        
        # Update metadata and save
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_image.shape[1],
            "width": out_image.shape[2],
            "transform": out_transform,
            "nodata": src.nodata
        })
        
        with rasterio.open(output_path, "w", **out_meta) as dest:
            dest.write(out_image)
            
        print(f"Clipping complete with buffer! Clipped GeoTIFF saved to: {output_path}")

# Print file sizes for comparison
original_size = os.path.getsize(geotiff_path) / (1024 * 1024)
clipped_size = os.path.getsize(output_path) / (1024 * 1024) if os.path.exists(output_path) else 0
print(f"\nFile sizes:")
print(f"Original GeoTIFF: {original_size:.2f} MB")
print(f"Clipped GeoTIFF: {clipped_size:.2f} MB")
print(f"Reduction: {(1 - clipped_size/original_size)*100:.1f}%")
```

```python
import os
import glob
import rasterio
from rasterio.mask import mask
import geopandas as gpd
from shapely.geometry import mapping
import numpy as np

# Define paths
input_folder = r"E:\InSAR Paper data\Sentinel data for the Gongmarri-Amdo section from 2018 to 2024\mm"
shapefile_path = r"E:\InSAR Paper data\Area shp 10km buffer\road_10km_buffer.shp"
output_folder = r"E:\InSAR Paper data\Time series clipped 10km"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Read the shapefile
print("Reading shapefile...")
try:
    shapefile = gpd.read_file(shapefile_path)
    print(f"Shapefile loaded successfully. CRS: {shapefile.crs}")
except Exception as e:
    print(f"Error reading shapefile: {e}")
    exit()

# Get all TIFF files in the input folder
tiff_files = glob.glob(os.path.join(input_folder, "*.tif"))
tiff_files.extend(glob.glob(os.path.join(input_folder, "*.tiff")))  # Also include .tiff extension

print(f"Found {len(tiff_files)} TIFF files to process")

# Process each TIFF file
for i, tiff_file in enumerate(tiff_files, 1):
    try:
        print(f"\nProcessing file {i}/{len(tiff_files)}: {os.path.basename(tiff_file)}")
        
        # Open the raster file
        with rasterio.open(tiff_file) as src:
            # Check if shapefile and raster have the same CRS, if not, reproject shapefile
            if shapefile.crs != src.crs:
                print(f"Reprojecting shapefile from {shapefile.crs} to {src.crs}")
                shapefile_reprojected = shapefile.to_crs(src.crs)
                geom = [mapping(shapefile_reprojected.geometry.unary_union)]
            else:
                geom = [mapping(shapefile.geometry.unary_union)]
            
            # Clip the raster with the shapefile
            try:
                out_image, out_transform = mask(src, geom, crop=True, nodata=src.nodata)
                
                # Update metadata for the clipped raster
                out_meta = src.meta.copy()
                out_meta.update({
                    "driver": "GTiff",
                    "height": out_image.shape[1],
                    "width": out_image.shape[2],
                    "transform": out_transform
                })
                
                # Generate output filename
                base_name = os.path.basename(tiff_file)
                name_without_ext = os.path.splitext(base_name)[0]
                output_filename = f"{name_without_ext}_clipped.tif"
                output_path = os.path.join(output_folder, output_filename)
                
                # Save the clipped raster
                with rasterio.open(output_path, "w", **out_meta) as dest:
                    dest.write(out_image)
                
                print(f"✓ Saved: {output_filename}")
                
            except Exception as e:
                print(f"✗ Error clipping {base_name}: {e}")
                
    except Exception as e:
        print(f"✗ Error processing {tiff_file}: {e}")

print(f"\nProcessing complete! Clipped files saved in: {output_folder}")
```

```python
import geopandas as gpd
from shapely.ops import unary_union
import os

# -----------------------------
# Paths
# -----------------------------
road_path = r"D:\1. Ph.D Research Work\STC-Data\Studyarea-road\road.shp"
study_area_path = r"E:\InSAR Paper data\new shp\New_Shapefile1.shp"
output_folder = r"E:\InSAR Paper data\road new"
output_path = os.path.join(output_folder, "road_clipped_clean.shp")

os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# Read files
# -----------------------------
roads = gpd.read_file(road_path)
study_area = gpd.read_file(study_area_path)

# -----------------------------
# Match CRS
# -----------------------------
if roads.crs != study_area.crs:
    study_area = study_area.to_crs(roads.crs)

# -----------------------------
# FIX 1 — Repair invalid geometries
# -----------------------------
roads["geometry"] = roads.buffer(0)
study_area["geometry"] = study_area.buffer(0)

# -----------------------------
# FIX 2 — Dissolve study area to single polygon
# -----------------------------
study_union = study_area.unary_union

# -----------------------------
# FIX 3 — Clip
# -----------------------------
clipped = roads.intersection(study_union)

clipped_gdf = gpd.GeoDataFrame(roads.drop(columns="geometry"),
                               geometry=clipped,
                               crs=roads.crs)

# -----------------------------
# FIX 4 — Remove empty & tiny fragments
# -----------------------------
clipped_gdf = clipped_gdf[~clipped_gdf.is_empty]
clipped_gdf = clipped_gdf[clipped_gdf.length > 5]  # remove <5 meters fragments

# -----------------------------
# FIX 5 — Explode multipart lines
# -----------------------------
clipped_gdf = clipped_gdf.explode(index_parts=False)

# -----------------------------
# FIX 6 — Snap precision (very important)
# -----------------------------
clipped_gdf["geometry"] = clipped_gdf.simplify(0.01)

# -----------------------------
# Save
# -----------------------------
clipped_gdf.to_file(output_path)

print("Clean clipping completed.")
print(f"Saved to: {output_path}")
```

# Conversion m to cm and mm for lower section tiffs

```python
import rasterio
import os
import glob
import numpy as np
from rasterio.plot import show
import warnings
warnings.filterwarnings('ignore')

def convert_m_to_mm(source_dir, dest_dir, scale_factor=1000.0):
    """
    Convert all GeoTIFF files in source_dir from meters to millimeters
    
    Parameters:
    source_dir: Source directory containing GeoTIFF files in meters
    dest_dir: Destination directory for converted files in millimeters
    scale_factor: Conversion factor (default 1000 for m to mm)
    """
    
    # Create destination directory
    os.makedirs(dest_dir, exist_ok=True)
    
    # Find all GeoTIFF files
    extensions = ['*.tif', '*.tiff', '*.TIF', '*.TIFF']
    geotiff_files = []
    for ext in extensions:
        geotiff_files.extend(glob.glob(os.path.join(source_dir, ext)))
    
    if not geotiff_files:
        print(f"No GeoTIFF files found in {source_dir}")
        return
    
    print(f"Found {len(geotiff_files)} GeoTIFF files")
    print("-" * 50)
    
    processed = 0
    failed = 0
    
    for file_path in sorted(geotiff_files):
        filename = os.path.basename(file_path)
        dest_path = os.path.join(dest_dir, filename)
        
        # Skip if already processed (optional)
        if os.path.exists(dest_path):
            print(f"⏭️  Skipping {filename} (already exists)")
            continue
            
        try:
            print(f"🔄 Processing: {filename}")
            
            with rasterio.open(file_path) as src:
                # Read metadata and data
                profile = src.profile
                data = src.read(1)
                
                # Handle NoData values if present
                nodata = src.nodata
                if nodata is not None:
                    # Create mask for valid data
                    valid_mask = ~np.isclose(data, nodata) if np.isnan(nodata) else (data != nodata)
                
                # Convert to millimeters (only where data is valid)
                data_mm = data.copy()
                if nodata is not None:
                    data_mm[valid_mask] = data[valid_mask] * scale_factor
                else:
                    data_mm = data * scale_factor
                
                # Update profile for output
                profile.update({
                    'dtype': rasterio.float32,
                    'driver': 'GTiff'
                })
                
                # Write output file
                with rasterio.open(dest_path, 'w', **profile) as dst:
                    dst.write(data_mm.astype(rasterio.float32), 1)
                    
                    # Copy metadata if needed
                    dst.update_tags(**src.tags())
                
                # Print statistics
                print(f"  ✓ Range: [{np.nanmin(data_mm):.2f}, {np.nanmax(data_mm):.2f}] mm")
                processed += 1
                
        except Exception as e:
            print(f"  ✗ Error: {str(e)}")
            failed += 1
    
    print("-" * 50)
    print(f"Summary: {processed} files processed, {failed} failed")

# Run the conversion
source_dir = r"E:\InSAR Paper data\def rate+cumul+summer+winter m(lower section)"
dest_dir = r"E:\InSAR Paper data\def rate+cumul+summer+winter mm(lower section)"

convert_m_to_mm(source_dir, dest_dir)
```

```python
import rasterio
import os
import numpy as np
from pathlib import Path

def convert_cm_to_mm(input_file, suffix='_mm', overwrite=False):
    """
    Convert a GeoTIFF from cm to mm
    
    Parameters:
    input_file: Path to input GeoTIFF (in cm)
    suffix: Suffix to add to output filename (default: '_mm')
    overwrite: Whether to overwrite existing output file (default: False)
    """
    
    # Validate input file
    if not os.path.exists(input_file):
        raise FileNotFoundError(f"Input file not found: {input_file}")
    
    # Generate output filename
    input_path = Path(input_file)
    output_path = input_path.parent / f"{input_path.stem}{suffix}{input_path.suffix}"
    
    # Check if output already exists
    if output_path.exists() and not overwrite:
        print(f"Output file already exists: {output_path}")
        response = input("Overwrite? (y/n): ").lower()
        if response != 'y':
            print("Conversion cancelled.")
            return None
    
    print(f"Converting: {input_path.name}")
    print(f"From: cm → To: mm")
    print("-" * 50)
    
    try:
        with rasterio.open(input_file) as src:
            # Read metadata
            profile = src.profile
            print(f"Dimensions: {src.width} x {src.height} pixels")
            print(f"CRS: {src.crs}")
            print(f"Bounds: {src.bounds}")
            
            # Read data
            data = src.read(1)
            nodata = src.nodata
            
            # Handle NoData values
            if nodata is not None:
                print(f"NoData value: {nodata}")
                # Create mask for valid pixels
                if np.isnan(nodata):
                    valid_mask = ~np.isnan(data)
                else:
                    valid_mask = data != nodata
                
                # Calculate statistics for valid pixels only
                valid_data_cm = data[valid_mask]
                if len(valid_data_cm) > 0:
                    print(f"Original range (cm): [{valid_data_cm.min():.2f}, {valid_data_cm.max():.2f}]")
                    print(f"Original mean (cm): {valid_data_cm.mean():.2f}")
            else:
                print("No NoData value specified")
                print(f"Original range (cm): [{data.min():.2f}, {data.max():.2f}]")
                print(f"Original mean (cm): {data.mean():.2f}")
                valid_mask = np.ones_like(data, dtype=bool)
            
            # Convert to mm (multiply by 10)
            data_mm = data.copy()
            data_mm[valid_mask] = data[valid_mask] * 10.0
            
            # Update profile
            profile.update({
                'dtype': rasterio.float32,
                'driver': 'GTiff',
                'compress': 'lzw'  # Add compression to save space
            })
            
            # Update NoData value if it exists
            if nodata is not None:
                if np.isnan(nodata):
                    profile.update({'nodata': np.nan})
                else:
                    profile.update({'nodata': nodata * 10.0})
            
            # Write output file
            with rasterio.open(output_path, 'w', **profile) as dst:
                dst.write(data_mm.astype(rasterio.float32), 1)
                
                # Copy and update metadata tags
                tags = src.tags()
                tags['CONVERSION'] = 'Converted from cm to mm'
                tags['ORIGINAL_UNITS'] = 'cm'
                tags['NEW_UNITS'] = 'mm'
                dst.update_tags(**tags)
            
            # Verify conversion
            if nodata is not None:
                valid_data_mm = data_mm[valid_mask]
                print(f"\nConverted range (mm): [{valid_data_mm.min():.2f}, {valid_data_mm.max():.2f}]")
                print(f"Converted mean (mm): {valid_data_mm.mean():.2f}")
            else:
                print(f"\nConverted range (mm): [{data_mm.min():.2f}, {data_mm.max():.2f}]")
                print(f"Converted mean (mm): {data_mm.mean():.2f}")
            
            print(f"\n✓ Successfully saved to: {output_path}")
            return output_path
            
    except Exception as e:
        print(f"✗ Error: {str(e)}")
        return None

# Run the conversion
if __name__ == "__main__":
    input_file = r"E:\InSAR Paper data\Sentinel data for the Gongmarri-Amdo section from 2018 to 2024\Deformation rate\deformation rate_clipped_cm.tif"
    
    # Convert with default settings
    output_file = convert_cm_to_mm(input_file)
    
    # If you want to force overwrite without asking:
    # output_file = convert_cm_to_mm(input_file, overwrite=True)
    
    # If you want a different suffix:
    # output_file = convert_cm_to_mm(input_file, suffix='_mm_converted')
```
